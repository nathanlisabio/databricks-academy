```python
bids = (spark.readStream
      .format("cloudFiles")
      .option("cloudFiles.format", "parquet")      # or csv, parquet, etc
      .option("cloudFiles.includeExistingFiles", "true")
      .option("cloudFiles.schemaLocation", "dbfs:/schemas/txlog_bids")  # REQUIRED
      .load("s3://dev-us-east-1.sabio.txlogs.parquet/bids/")             # the root path above yyyy/mm/dd/hh
)
```

```python
bids.printSchema()
```

```python
display(bids)
```

```python
s3_checkpoint_path = "s3://dev-databricks-appscidev/checkpoints/txlog_bids"
(bids
    .writeStream
    # .format("delta")
    .option("checkpointLocation", s3_checkpoint_path)
    # .outputMode("append")
    .trigger(availableNow=True)
    .table("dbacademy.silver.txlog_bids")
)
```
